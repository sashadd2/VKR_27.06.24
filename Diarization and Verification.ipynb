```python
def get_all_speakers(rttm_path):
    with open(rttm_path, 'r') as file:
        lines = file.readlines()
    speakers = set()
    for i, line in enumerate(lines):
        parts = line.strip().split()
        speakers.add(parts[7])
    return speakers
```
